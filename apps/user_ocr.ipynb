```python
from PIL import Image
```

```python
import cv2
```

```python
Image.open("receipt1.jpg")
```

```python
cv2.im
```

```python
image = cv2.imread("receipt1.jpg") # gets np array, represents image
```

```python
cv2.imshow("original",image)
```

```python
image.shape # 448 rows, 253 columns, 3 channels (RGB)
# BUT 253 pixels wide, 448 pixels tall
```

```python
# resize image

r = 100.0 / image.shape[1]
dim = (100, int(image.shape[0] * r))

resized = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
cv2.imshow("resized",resized)
```

```python
# flip image upsidedown

(h,w) = image.shape[:2]
center = (w/2, h/2)
M = cv2.getRotationMatrix2D(center, 180, 1.0)
rotated = cv2.warpAffine(image, M, (w,h))
cv2.imshow("rotated",rotated)
```

```python
image.shape
```

```python
# crop image

cropped = image[70:170,100:200] #startY:endY, startX:endX
cv2.imshow("cropped",cropped)
```

```python
# save cropped image
cv2.imwrite("cropped.png",cropped)
```

```python
# restore image from array, for pytesseract
img_crp = Image.fromarray(cropped)
```
